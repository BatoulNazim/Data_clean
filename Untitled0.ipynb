```python

import pandas as pd
import random

# إنشاء Dataset وهمي
num_customers = 50

# توليد بيانات العملاء
customer_ids = [f"C{i+1}" for i in range(num_customers)]
purchase_frequency = [random.randint(1, 20) for _ in range(num_customers)]
total_purchase_amount = [round(random.uniform(50, 2000), 2) for _ in range(num_customers)]

# توليد قائمة منتجات لكل عميل
products = ["Bread", "Butter", "Milk", "Eggs", "Laptop", "Mouse", "Keyboard", "Phone", "Charger", "Juice"]
items_purchased = [random.sample(products, random.randint(2, 5)) for _ in range(num_customers)]

# إنشاء DataFrame
data = pd.DataFrame({
    "CustomerID": customer_ids,
    "PurchaseFrequency": purchase_frequency,
    "TotalPurchaseAmount": total_purchase_amount,
    "ItemsPurchased": items_purchased
})

# عرض أول 5 صفوف للتأكد من البيانات
data.head()
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# اختيار الأعمدة الخاصة بالتحليل
X = data[["PurchaseFrequency", "TotalPurchaseAmount"]]

# إنشاء نموذج K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
data["Cluster"] = kmeans.fit_predict(X)

# رسم المجموعات
plt.figure(figsize=(8,6))
plt.scatter(data["PurchaseFrequency"], data["TotalPurchaseAmount"], c=data["Cluster"], cmap="viridis")
plt.xlabel("Purchase Frequency")
plt.ylabel("Total Purchase Amount")
plt.title("Customer Segmentation (K-Means)")
plt.colorbar(label="Cluster")
plt.show()
```

```python

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# تجهيز البيانات لقواعد الربط
transactions = data["ItemsPurchased"].tolist()

# تحويل البيانات لصيغة مناسبة
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# تطبيق Apriori لاستخراج المجموعات المتكررة
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

# استخراج القواعد
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# عرض أفضل 3 قواعد بناءً على Lift
top_rules = rules.sort_values(by="lift", ascending=False).head(3)
top_rules[["antecedents", "consequents", "support", "confidence", "lift"]]
```
